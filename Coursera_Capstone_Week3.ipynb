## Read Data

```python
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/shayy07/Coursera_Capstone/master/TORONTONB.csv")
df.head()
```

## Process Borough

Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.<br>
So we need to remove rows with Borough = "Not assigned"

```python
#Read data
df_test=df.sort_values(by=['Postcode'])
df_test['Borough'].value_counts()
```

```python
df_test=df_test[df_test.Borough!="Not assigned"]
df_test['Borough'].value_counts()
```

# Process Neighbourhood

Rename neighbourhood as borough when neighbourhood is "Not assigned" but borough isn't.

```python
df_test['Neighbourhood'][(df_test['Neighbourhood'] == "Not assigned")]
df_test['Neighbourhood'][(df_test['Neighbourhood'] == "Not assigned")] = df_test['Borough']
df_test['Neighbourhood'][(df_test['Neighbourhood'] == "Not assigned")]
```

# Merge Neighbourhood with Same Postcode

```python
df_test["NBtmp"]=df_test["Neighbourhood"]+","
df_clean = df_test.groupby(by=['Postcode','Borough'])['NBtmp'].sum().reset_index(name ='Neighbourhood')
df_clean
```

# Add Geocodes

```python
df_geo=pd.read_csv("https://raw.githubusercontent.com/shayy07/Coursera_Capstone/master/Geospatial_Coordinates.csv")
df_geo
result = pd.merge(df_clean, df_geo, how='left', on='Postcode')
result
```
